```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Dataset=pd.read_csv('parkinsons.csv')
Dataset
```

```python
Dataset.drop(['name'],axis=1,inplace=True)
Dataset
```

```python
Dataset.replace('Missing',np.nan,inplace=True)
Dataset
```

```python
Dataset.dropna(inplace=True)
Dataset
```

```python
Dataset.replace('None',np.nan,inplace=True)
Dataset
```

```python
Dataset.dropna(inplace=True)
Dataset
```

```python
Dataset['status'].replace("Parkinson's",1,inplace=True)
Dataset['status'].replace("Healthy",0,inplace=True)
Dataset
```

```python
Y=Dataset.iloc[:,16].values
Dataset.drop(['MDVP:Jitter(Abs)'],axis=1,inplace=True)
Dataset
Dataset.drop(['status'],axis=1,inplace=True)
X=Dataset.iloc[:,:].values
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_test = sc.fit_transform(X_test)
X_train = sc.fit_transform(X_train)
print(X_test)
```

```python
#random forest  
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
Y_pred = classifier.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
print(accuracy_score(Y_test, Y_pred))
```
